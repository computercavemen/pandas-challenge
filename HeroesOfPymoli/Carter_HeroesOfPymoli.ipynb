```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# count unique players & verify numbers 
total_players = len(pd.unique(purchase_data_df["SN"]))
total_players
#total_players = len(purchase_data_df["SN"].value_counts())
#total_players
```

```python
# add column to dataframe with lists total players
pd.DataFrame({"Total Players": [total_players]})
```

```python
# Purchasing analysis calculations & variables

# unique items
unique_items = len(purchase_data_df["Item ID"].value_counts())

# average price
price = round(purchase_data_df["Price"].mean(),2)
average_price = "${:,.2f}". format(price)

# number of purchases
purchases = len(purchase_data_df["Purchase ID"].value_counts())

# total revenue
revenue = round(purchase_data_df["Price"].sum(),2)
total_revenue= "${:,.2f}". format(revenue)

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# creating a dataframe by item information  
popular_purchase = purchase_data_df.groupby(["Item ID", "Item Name"])
popular_purchase_count = popular_purchase["Price"].count()
popular_average = round(popular_purchase["Price"].mean(), 2)
popular_total = round(popular_purchase["Price"].sum(), 2)
```

```python
# create dataframe
popular_items_df = pd.DataFrame({"Purchase Count": popular_purchase_count,
              "Item Price": popular_average,
              "Total Purchase Value": popular_total
             })
popular_items_df
```

```python
popular_items_df.sort_values(["Purchase Count"], axis=0,
                 ascending = False, inplace=True)
popular_items_df.head()
```
